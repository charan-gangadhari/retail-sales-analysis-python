```python
import pandas as pd

# Create dataset manually
data = {
    'Category': ['Technology','Furniture','Office Supplies','Technology','Furniture','Office Supplies','Technology','Furniture'],
    'Product': ['Laptop','Chair','Pen','Phone','Table','Paper','Monitor','Sofa'],
    'Region': ['East','West','South','North','East','West','South','North'],
    'Sales': [1200, 300, 50, 800, 450, 100, 600, 700],
    'Order Date': ['2023-01-10','2023-02-15','2023-03-12','2023-04-18','2023-05-20','2023-06-25','2023-07-30','2023-08-05']
}

df = pd.DataFrame(data)

df.head()
```

```python
# Convert date
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Total Sales
print("Total Sales:", df['Sales'].sum())

# Sales by Category
print(df.groupby('Category')['Sales'].sum())

# Sales by Region
print(df.groupby('Region')['Sales'].sum())
```

```python
import matplotlib.pyplot as plt

df.groupby('Category')['Sales'].sum().plot(kind='bar', title='Sales by Category')
plt.show()
```

```python
# Monthly trend
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line', title='Monthly Sales Trend')
plt.show()
```
